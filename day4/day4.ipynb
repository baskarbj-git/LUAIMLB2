# Question 2 :
Research on range() functions and its parameters. Create a markdown cell and write in your own words
(no copy-paste from google please) what you understand about it. Implement a small program of your
choice on the same

Research Outcome: Range function generates immutable integer sequences based on input arguments. It is applicable on integers. 
Output generated by range can be used in for, while loops as well as in constructing other containers like list/set/tuple

Below are the 3 differnet usage scenarios,
1. Generate sequence upto an end value, i.e., range(N) generates numbers 0,1...N-1
2. Generate sequence between start, end value. i.e., range(n,m) generates n,n+1....m-1
3. Generate sequence between start, end value with a step value. i.e., as in '2' but increments in steps "k" n,n+k, n+2k..,
   The step value can also be negative, where it will be consider as decrementing from start value to end value

```python
type(range(5))

for i in range(5):
    print(1)

print(list(range(5)))
print(set(range(5)))
print(tuple(range(5)))

print(list(range(5,10)))
print(list(range(5,10,2)))
print(list(range(5,10,-1)))
print(list(range(50,10,-5)))
```

# Question 3 
Consider two numbers. Perform their subtraction and if the result of subtraction is greater than 25, print
their multiplication result else print their division result.

```python
#Since order not mentioned, taking abs diff, a * b, a/b
a=int(input("First Number"))
b=int(input("Second Number"))

diff = abs(a-b)
if diff > 25:
    print(a*b)
else:
    print(a/b)
```

# Question 4 
Consider a list of 10 elements of integer values. If the number in the list is divisible by 2, print the result as
"square of that number minus 2".

```python
elements = [23,45,67,12,89,233,3,56,69,90]

result = [(n ** 2) - 2 for n in elements if (n%2) == 0]
print(result)
```
